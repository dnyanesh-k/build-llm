### Tensor Indexing

```python
import numpy as np
import torch
```

### Create a Rank1 tensor

```python
vector = torch.tensor([10, 20, 30, 40], dtype=torch.float32)
vector
```

### Positive Indexing

```python
print(f"vector[0] = {vector[0]}")
print(f"vector[1] = {vector[1]}")
print(f"vector[len(vector)-1] = {vector[len(vector)-1]}")
```

### Negative Indexing

```python
print(f"vector[-1] = {vector[-1]}")
print(f"vector[-2] = {vector[-2]}")
print(f"vector[-len(vector)] = {vector[-len(vector)]}")
```

### create rank2 tensor

```python
matrix = torch.tensor([[10, 20], [30, 40], [50, 60]])
matrix
```

### Positive Indexing

```python
print(f"first row matrix = {matrix[0]}")
print(f"matrix [0][0] = {matrix[0][0]}")
print(f"matrix[0][1] = {matrix[0][1]}")
```

```python
print(f"the last row matrix[-1] = {matrix[-1]}")
print(f"matrix [0,1] = {matrix[0,1]}")
print(f"matrix[0][1] = {matrix[0][1]}")
```
