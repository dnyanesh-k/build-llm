### Tensor Slicing

```python
import numpy as np
import torch
```

```python
vector = torch.tensor([10, 20, 30, 40, 50, 60])
vector
```

```python
# start =2 , stop= 5 (excluded)
print(f"vector[2:5] = {vector[2:5]}")

# start= 2, stop = 4(excluded)
print(f"vector[2:4] = {vector[2:4]}")

# start = 0, stop = 4 (excluded)
print(f"vector[:4] = {vector[:4]}")

# start = 3, stop = 6 (excluded)
print(f"vector[3:] = {vector[3:]}")

# get all the values
print(f"vector[0:6] = {vector[0:6]}")

# get all the values
print(f"vector[0:] = {vector[0:]}")

# get all the values
print(f"vector[:6] = {vector[:6]}")

# get all the values
print(f"vector[:] = {vector[:]}")
```

```python
# get alternate values
print(f"vector[::2] = {vector[::2]}")

# get 3rd values
print(f"vector[::3] = {vector[::3]}")

# get all values in reverse order
# print(f"vector[::-1] = {vector[::-1]}")
```
