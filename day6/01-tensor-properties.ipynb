### Tensor proeprties

```python
import numpy as np
import torch
```

### create a rank1 tensor without autograd

```python
vector = torch.tensor([10, 20, 30, 40, 50], dtype=torch.float32)
vector
```

```python
print(f"Data type of every value in tensor : {vector.dtype}")
print(f"Size of tensor = {vector.numel()} or {vector.size()}")
print(f"Shape of tensor = {vector.shape}")
print(f"No of Dimensions = {vector.ndim} or {vector.dim()}")
print(f"Thge divice on which vector is going to perform ops = {vector.device}")
print(f"Is autograd enabled = {vector.requires_grad}")
print(f"Is thios tensor a leaf node of DAG = {vector.is_leaf}")
```

### Create a rank 2 tensor without autograd

```python
# rank 2 tensor
matrix = torch.tensor([[10, 20], [30, 40], [50, 60]], dtype=torch.float32)
```

```python
print(f"Data type = {matrix.dtype}")
print(f"Size of tensor = {matrix.size()} or {matrix.numel()}")
print(f"Shape of tensor = {matrix.shape}")
print(f"No of dim = {matrix.ndim} or {matrix.dim()}")
print(f"Device = {matrix.device}")
print(f"Is autograd enabled = {matrix.requires_grad}")
print(f"Is leaf node of DAG = {matrix.is_leaf}")
```
