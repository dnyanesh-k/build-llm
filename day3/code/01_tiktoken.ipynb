```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding('gpt2')
```

```python

corpus = "I love machine learning. Welcome  to build your own llm"
```

```python
# encode the corpus to get the tokens
tokens = tokenizer.encode(corpus)
tokens
```

```python
# get the original corpus using the tokens
original_corpus = tokenizer.decode(tokens=tokens)
original_corpus
```

```python
# size of tiktoken vocabulary
print(f"Size of tiktoken Vocab = {tokenizer.n_vocab}")
```

```python
for token in tokens:
    print(f"{token} : {tokenizer.decode([token])}")
```

```python
# create a tokenizer using gpt4
tokenizer_4o = tiktoken.get_encoding('o200k_base')

tokenizer_4o.n_vocab
```

```python
tokens = tokenizer_4o.encode(corpus)
tokens
```
