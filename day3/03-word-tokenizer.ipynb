word tokenizer

```python
import numpy as np
import re
```

```python
class WordTokenizer:
    def __init__(self, corpus):
        # split the all words
        all_words = re.split(r'[,.?"\'! ]', corpus)

        #  get the final list of words
        words = []
        for word in all_words:
            if word.strip():
                words.append(word.lower())

        # add the special tokens
        words.extend(['[UNK]','<|im_start|>', '<|im_end|>'])

        # build the dictionaries
        self.encoding_dictionary = {word:token_id for token_id, word in enumerate(words)}
        self.decoding_dictionary = {token_id:word for token_id, word in enumerate(words)}

    def encode(self, text):
        words = re.split(r'[,.?"\'?! ]', text)
        return [self.encoding_dictionary.get(word.lower()) if self.encoding_dictionary.get(word.lower()) else self.encoding_dictionary.get('[UNK]') for word in words]
    
    def decode(self, tokens):
        return [self.decoding_dictionary[token] for token in tokens]
    
    def n_vocab(self):
        return len(self.encoding_dictionary.items())
```

```python
file_name = 'data.txt'

with open(file_name, 'r') as file:
    corpus = file.read()    
```

Save the tokenizer to a file

# **Pickling**
- Normally, when you close a Python program, everything you created—like your dictionary of words or a list of numbers—disappears from memory. If you want them back, you have to run the code to recreate them. 
- Pickling lets you take that object, turn it into a stream of bytes, and save it as a file (.pkl). 
- Later, you can unpickle it to bring the exact same object back to life, exactly where you left off.

## Why use it?
- **Saving Progress:** If it took your computer 10 minutes to process a huge list of words into a dictionary, you don't want to wait 10 minutes every time you run your script. 
- **Sending Data:** You can pickle a complex object and email it to a friend. They can "unpickle" it on their computer and have the exact same object you had. 
- **Consistency:** It ensures your "Word-to-ID" mappings never change

```python
# pickle: A Python library used to "freeze" (serialize) objects into a byte format.
import pickle

# wb = Stands for Write Binary. It tells Python you are creating a file and saving data in a non-text format
with open('tokenizer.pkl','wb') as file: 
    pickle.dump(tokenizer, file)  # This actually takes your tokenizer (including its encoding_dictionary and all its logic) and writes it into the file.
```
