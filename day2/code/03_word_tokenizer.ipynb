# Word level tokenizer

```python
class WordTokenizer():
    def __init__(self, vocab):
        # persist the vocabulary
        self.vocab = vocab

        # build the dictionaries
        self.stoi = {ch: id for id, ch in enumerate(vocab)}
        self.itos = {id: ch for id, ch in enumerate(vocab)}

    def encode(self, words):
        # convert every character to corresponding integer value
        return [self.stoi[word] for word in words]

    def decode(self, encoded_value):
        # convert every integer into  corresponding character
        return " ".join([self.itos[value] for value in encoded_value])

    def print_info(self):
        print(self.stoi)
        print(self.itos)
        
```

```python
corpus = "I love machine learning. I love Artificial Intelligence. welcome to sunbeam. I have ped dog.merry christmas."
# find the words from corpus
words = corpus.split()
words
```

```python
import re

# find the words from the corpus
words = re.split(r'[,.?|!"\' ]', corpus)

# remove the empty words
words = [word.strip() for word in words if word.strip()]

# find unique words and sort them
words = sorted(set(words))
words
```

```python
tokenizer = WordTokenizer(words)

tokenizer.print_info()
```

Test the Tokenizer

```python
new_text = "welcome to machine Intelligence"
new_text_words = list(re.split(r'[,.?|!"\' ]', new_text))
new_text_words = [word.strip() for word in new_text_words if word.strip()]

encoded_value = tokenizer.encode(new_text_words)
encoded_value
```
