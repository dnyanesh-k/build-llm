# Character Level Tokenizer

```python
class CharacterTokenizer():
    def __init__(self, vocab):
        # persist the vocab
        self.vocab = vocab

        # build the dictionaries
        self.stoi = {ch: i for i, ch in enumerate(vocab)}
        self.itos = {i: ch for i, ch in enumerate(vocab)}

    def encode(self, text):
        # convert every character to corresponding integer value
        return [self.stoi[ch] for ch in text]

    def decode(self, encoded_value):
        # convert every integer into corresponding character
        return "".join([self.itos[num] for num in encoded_value])
    
    def print_info(self):
        print(self.stoi)
        print(self.itos)
       
```

```python
corpus = "I love machine learning. I love Artificial Intelligence. Welcome to sunbeam. I have a ped dog. Merry Christmas"

# get all the unique characters
vocab = sorted(set(list(corpus)))

tokenizer = CharacterTokenizer(vocab)

tokenizer.print_info()
```

```python
new_text = "Welcome to build your llm from scratch"

encoded_value = tokenizer.encode(new_text)
encoded_value
```

```python
# decode the tokens to thier original value
tokenizer.decode(encoded_value=encoded_value)
```
