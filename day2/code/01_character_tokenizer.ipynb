Character Tokenizer

```python
import numpy as np
```

get the data

```python
corpus = "I love machine learning. I love artificial intelligence"
```

tokenization logic

```python
# build the dictionary to hold the token ids of ebery character in corpus
stoi = {}

# build the dictionary to hold the tokens for every token ids
itos = {}
```

Build the vocabulory (Collection of Unique characters)

```python
vocab = sorted(set(list(corpus)))
vocab
```

```python
token_id = 0
for ch in vocab:
    print(f"char = {ch} ; token_id = {token_id}")
    token_id += 1
```

```python
for token_id, ch in enumerate(vocab):
    # get integer representation of every token and add the mapping to the dictionary
    stoi[ch] = token_id

    # get the character representation of every token id and add mapping to the dictionary
    itos[token_id] = ch
# stoi
itos    
```

```python
# get the integer representation of every character and add mapping to teh dictionary
stoi = {ch: token_id for token_id, ch in enumerate(vocab)}

# get the character representation of every token and add mapping to the dictionary
itos = {token_id: ch for token_id, ch in enumerate(vocab)}
# itos
stoi
```
