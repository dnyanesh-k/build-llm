Character Level Tokenizer

```python
class CharacterTokenizer:
    def __init__(self, vocab):
        # persist the vocabulary
        self.vocab = vocab

        # build the dictionaries
        self.stoi = {ch:i for i, ch in enumerate(vocab)}
        self.itos = {i:ch for i, ch in enumerate(vocab)}

    def encode(self, text):
        # convert every character to corresponding integer (token id) value
        return [self.stoi[ch] for ch in text]

    def decode(self, encoded_value):
        # convert every integer to corresponding character
        return "".join([self.itos[num] for num in encoded_value])

    def print_info(self):
        print(self.stoi)
        print(self.itos)    
```

```python
corpus = "I love machine learning, I love artificial intelligence. I love content creation and also like to do gym"

# get all unique characters
vocabulary = sorted(set(list(corpus)))

# create Tokenizer
tokenizer = CharacterTokenizer(vocabulary)

tokenizer.print_info()
```

```python
# new text to verify the tokenization
new_text = "I like to go to gym"
encoded_value = tokenizer.encode(new_text)
encoded_value
```
