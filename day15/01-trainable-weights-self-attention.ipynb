## Trainable weights self attention msm

```python
import numpy as np
import torch
import math
import torch.nn as nn
```

### def the self attention class

```python
class SelfAttentionV1(nn.Module):
    # initializer
    def __init__(self, dimension_input, dimension_output):
        # initialize the super class
        super().__init__()

        # Create the weight metrices for Q, K and V
        self.weight_query = torch.nn.Parameter(torch.rand(dimension_input, dimension_output))
        self.weight_key = torch.nn.Parameter(torch.rand(dimension_input, dimension_output))
        self.weight_value = torch.nn.Parameter(torch.rand(dimension_input, dimension_output))

     # will be called when data is passed to the model
    def forward(self, inputs):
        # create queries, keys and values
        queries = inputs @ self.weight_query
        keys = inputs @ self.weight_key
        values = inputs @ self.weight_value

        # calculate attention scores
        attention_scores = queries @ keys.T

        # scale attention scores
        dimension_keys = inputs.shape[-1]
        scaled_weights = attention_scores / math.sqrt(dimension_keys)
  
        # calculate attention weights
        attention_weights = torch.softmax(scaled_weights, dim=-1)

        # calculate context vector
        context_vector = attention_weights @ values

        return context_vector 
```

```python
# test the class
inputs = torch.tensor([
   [0.43, 0.15, 0.89], # Your 
   [0.55, 0.87, 0.66], # journey
   [0.57, 0.85, 0.64], # starts
   [0.22, 0.58, 0.33], # with
   [0.77, 0.25, 0.10], # one
   [0.05, 0.80, 0.55]  # step
])
```

```python
# set the seed value to get same output every time
torch.manual_seed(123)

# input dimensions
dimension_input = inputs.shape[1]

dimention_output = 2
```

```python
# create the class object
self_attention = SelfAttentionV1(dimension_input=dimension_input, dimension_output=dimention_output)
```

```python
# calculate context vector
context_vector = self_attention(inputs)
context_vector
```
