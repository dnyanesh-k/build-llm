## Simple attention calculation

```python
import numpy as np
import torch
import gensim.downloader as api
import tiktoken
```

### get the input

```python
sentence = "the animal was tired and did not cross the road"
```

### get the tokenizer

```python
tokenizer = tiktoken.get_encoding("gpt2")

# create the tokens
token_ids = tokenizer.encode(sentence)
token_ids
```

### get the embedding model

```python
model = api.load("word2vec-google-news-300")
```

```python
# create the embeddings for 300 dimensions
input_embeddings = torch.tensor(model[token_ids])
input_embeddings.shape
```

### calculate the attention scores

```python
# store all the attention scores
attention_scores = torch.empty((input_embeddings.shape[0], input_embeddings.shape[0]))
attention_scores
```

```python
# find the attention wrt the word animal
query = input_embeddings[1]

# find the ateention of every embedding in input_embeddings wrt to every other embedding
for index, embedding in enumerate(input_embeddings):
    # calculate the attention
    attention_score = torch.dot(query, embedding)
    print(f"Attention score = {attention_score}")
```

```python
# get all embeddings from input_embeddings
for query_index, query in enumerate(input_embeddings):
    # find the attention score for this query wrt all others
    query_attention_scores = torch.empty(input_embeddings.shape[0])

    # find attention wrt all embeddings from input_embeddings
    for index, embedding in enumerate(input_embeddings):
        # find the attention score
        query_attention_scores[index] = torch.dot(query, embedding)

    # append the attention score of query to main to main attention_scores
    attention_scores[query_index] = query_attention_scores
```

```python
# print all attention scores
attention_scores
```
