## Tensor Creation

```python
import numpy as np 
import torch
```

Using python data

```python
# using python list
# scalar = torch.tensor(10)
vector = torch.tensor([10, 20, 30, 40])
matrix = torch.tensor([[10, 20], [30, 40]])
matrix
```

```python
# using python tuple
vector_using_tuple = torch.tensor((10, 20, 30, 40))
vector_using_tuple
```

### using numpy array

```python
array1 = np.array([10, 20, 30, 40])
vector_using_np_array = torch.tensor(array1)
vector_using_np_array
```

```python
matrix_using_np_array =np.array([[10, 20], [30, 40]])
matrix_using_np_array
```
