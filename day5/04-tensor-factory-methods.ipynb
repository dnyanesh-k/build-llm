## tensor creation using factory methods


```python
import numpy as np
import torch
```

### empty tensors

```python
empty_tensor = torch.empty(2, 2)
empty_tensor
```

### empty tensor copy

```python
empty_tensor_copy = torch.empty_like(empty_tensor)
empty_tensor_copy
```

### zero tensor

```python
zeroed_tensor = torch.zeros(3, 3)
zeroed_tensor
```

### Ones

```python
ones_tensor = torch.ones(2, 3)
ones_tensor
```

### Full tensor

```python
full_tensor = torch.full((3, 3), 57)
full_tensor
```

### Using arange

```python
sample_tensor = torch.arange(0, 10)
sample_tensor
```

### Identity Tensor

```python
identity_tensor = torch.eye(5)
identity_tensor
```

### diagonal tensor

```python
diagonal_tensor = torch.diag(torch.tensor([0, 1, 2, 3, 4, 5]))
diagonal_tensor
```

### Using random integers

```python
random_int_tensor = torch.randint(11, 13, (4, 4))
random_int_tensor
```

### Using random permutation

```python
per_tensor = torch.randperm(10)
per_tensor
```
