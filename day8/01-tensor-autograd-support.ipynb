### Tensor autograd support [docs](https://docs.pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html)

**Autograd** is PyTorch's automatic differentiation engine.
- `torch.autograd` is PyTorch’s automatic differentiation engine that powers neural network training. 
- It is the "brain" that allows a neural network to learn by automatically calculating the gradients (derivatives) needed to update its weights

### How It Works?
- When you perform operations on tensors, Autograd builds a dynamic computational graph in the background. 
- This graph tracks every step taken to reach an output (like your loss function)

**Tracking:** If a tensor has `requires_grad=True`, Autograd records every operation performed on it.

**The Graph:** It creates a **Directed Acyclic Graph (DAG)** of "Function" objects. - The leaves are your input tensors, and the root is your final output.

**Backward Pass:** When you call `.backward()` on your final output (the loss), Autograd traverses the graph in reverse. 
- It uses the chain rule to compute how much each input tensor contributed to that final output.

**Storing Gradients:** The resulting gradients are stored in each tensor’s `.grad` attribute

```python
# define a function in terms of y
y_hat = scalar ** 4
y_hat
```

### create a tensor with autograd support

```python
scalar = torch.tensor(10.0, requires_grad=True)
scalar
```

```python
scalar.grad
```

```python
y_hat = scalar ** 4
y_hat
```

```python
# backward prop on y_hat
y_hat.backward()
```

```python
# find the derivative of y_hat
scalar.grad
```

### create a new tensor with autograd

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
x.grad
```

```python
# y = f(x)
y = (x ** 2) + (3 * x)
y
```

```python
# calculate derivative
def calculate_derivative(x):
    return (x * 2) + (3)
```

```python
calculate_derivative(2.0)
```

```python
y.backward()
```

```python
x.grad
```
