# One Hot Encoding
- Because in the entire list of numbers, only one position is "hot" (set to 1), and all others are "cold" (0).

## Why do we need this?
- Imagine if you just gave the computer the ID numbers:
`Cat = 1`
`Dog = 2`
`Wolf = 100`
- A machine learning model might think a Wolf is *"bigger"* or *"more important"* than a Cat because 100 is bigger than 1. 
Or it might think a `Cat + Dog = a different word`.
- One-Hot Encoding fixes this. 
- It puts every word in its own separate *"lane."* 
- Because every vector has exactly one 1 and the rest 0, the computer treats them all as equal but different categories.

```python
def one_hot_encoding(word, vocab_size):
    # tensor vector
    # vector = torch.zeros(vocab_size)

    # array vector
    vector = np.zeros(vocab_size)

    # set the index value of word to 1
    index = dictionary[word]
    vector[index] = 1
    return vector
```

```python
for word in words:
    print(f"Embedding of {word} = {one_hot_encoding(word, len(words))}")
```
