### Tokenization

```python
import numpy as np
import torch
import re
import tiktoken
```

```python
with open('data.txt', 'r') as file:
    raw_text = file.read()
```

```python
raw_text[:100]
```

### create tokenizer

```python
class WordTokenizer():
    def __init__(self, raw_text):
        words = re.split(r'[,.?|!"\' ]', raw_text)

        # remove the empty words
        words = [word.strip() for word in words if word.strip()]

        # find the unique words and sort them  
        self.vocab = sorted(list(set(words)))

        # build the dictionaries
        self.stoi = {ch:i for i, ch in enumerate(self.vocab)}
        self.itos = {i:ch for i, ch in enumerate(self.vocab)}

    def encode(self, words):
        return [self.stoi[word] for word in words]

    def decode(self, encoded_values):
        return " ".join([self.itos[num] for num in encoded_values])
    
    def print_info(self):
        print(stoi)
        print(itos)
        
    def n_vocab(self):
        return len(self.vocab)
```

```python
tokenizer = WordTokenizer(raw_text=raw_text)
tokenizer.n_vocab()
```

```python
# get the tokens for 1st 10 words
words = re.split(r'[,.?|!"\' ]',raw_text)
words = [word.strip() for word in words if word.strip()]
stoi = tokenizer.encode(words)
tokens = stoi[:10]
```

```python
# decode these tokens
tokenizer.decode(tokens[:10])
```

### Create `tiktoken` to create token

```python
tokenizer = tiktoken.get_encoding('gpt2')
print(tokenizer.n_vocab)
```

```python
tokens = tokenizer.encode(raw_text)
tokens[:10]
```

```python
# decode the tokens
original_text = tokenizer.decode(tokens[:10])
original_text
```
